```python
# To get column names to first row
def column_name_to_first_row(df):
    # Find where does data start
    data_start=0
    for i, data in enumerate(df.iloc[:,0]):
        data_start=i
        if str(data).isnumeric():
            break
    # To get column names to first row from last non null row before data starts
    no_of_columns=len(df.columns)
    for j in range(no_of_columns):
        df_temp=df.iloc[0:data_start,[j]]
        i=data_start-1
        while i>=0:
            if not pd.isnull(df_temp.iloc[i,0]):
                df.iloc[0,j]=df_temp.iloc[i,0]
                break
            i-=1

    # State data start
    state_data_start=0
    for i, data in enumerate(df.iloc[:,0]):
        state_data_start=i
        if str(data).lower().find("state")>-1:
            break
    # Remove unwanted rows
    df.drop(df.index[1:state_data_start],inplace=True)
    return df
```

```python
def check_name_of_column_in_match(column_name,match):
    return all(x in column_name for x in match)
```

```python
#column names to search
matches_S_no = ["s.","no"]
matches_district = ["district"]
matches_murder = ["murder", "302"]
# matches_death = ["death","negligence","304"] #remove
matches_riot = ["riot","total"] #to do
matches_dacoity= ["dacoity","total"] #to do
matches_robbery= ["robbery","392"] #to do
matches_dowry_death= ["death","dowry"]
matches_theft= ["theft","total"]
matches_kidnap= ["kidnap","total"]
matches_rape= ["rape"]
 # if not check_name_of_column(col_name,matches_not_contain_rape) and check_name_of_column(col_name,matches_rape):
matches_not_contain_rape= ["attempt"]
# matches_fraud= ["fraud","total"]  #remove
# matches_rash_driving= ["rash","driving","total"]  #remove
matches_assault_on_women= ["assault","women","total"]
matches_total_crimes= ["crimes","total"]
matches_not_contain_total_crimes= ["miscellaneous"]
# if not check_name_of_column(col_name,matches_not_contain_total_crimes) and check_name_of_column(col_name,matches_total_crimes):
matches_not_contain_space=[" "]
```

```python
def rename_and_remove_unwanted_column(df):
    no_of_columns=len(df.columns)
    wanted_columns=[]

# To rename column
    for j in range(no_of_columns):
        col_name=df.iloc[0,j].lower()
        if check_name_of_column_in_match(col_name,matches_district):
            df = df.rename(columns={df.columns[j]: 'District'})
            wanted_columns.append('District')
        elif check_name_of_column_in_match(col_name,matches_S_no):
            df = df.rename(columns={df.columns[j]: 'S. No.'})
            wanted_columns.append('S. No.')   
        elif check_name_of_column_in_match(col_name,matches_murder):
            df = df.rename(columns={df.columns[j]: 'Murder'})
            wanted_columns.append('Murder')
        elif check_name_of_column_in_match(col_name,matches_riot):
            df = df.rename(columns={df.columns[j]: 'Riot'})
            wanted_columns.append('Riot')
        elif check_name_of_column_in_match(col_name,matches_dacoity):
            df = df.rename(columns={df.columns[j]: 'Dacoity'})
            wanted_columns.append('Dacoity')
        elif check_name_of_column_in_match(col_name,matches_robbery):
            df = df.rename(columns={df.columns[j]: 'Robbery'})
            wanted_columns.append('Robbery')
        # elif check_name_of_column_in_match(col_name,matches_death):
        #     df = df.rename(columns={df.columns[j]: 'Death by negligence'})
        #     wanted_columns.append('Death by negligence')
        elif check_name_of_column_in_match(col_name,matches_dowry_death):
            df = df.rename(columns={df.columns[j]: 'Dowry death'})
            wanted_columns.append('Dowry death')
        elif check_name_of_column_in_match(col_name,matches_theft):
            df = df.rename(columns={df.columns[j]: 'Theft'})
            wanted_columns.append('Theft')
        elif check_name_of_column_in_match(col_name,matches_kidnap):
            df = df.rename(columns={df.columns[j]: 'Kidnapping and abduction'})
            wanted_columns.append('Kidnapping and abduction')
        elif not check_name_of_column_in_match(col_name,matches_not_contain_rape) and check_name_of_column_in_match(col_name,matches_rape):
            df = df.rename(columns={df.columns[j]: 'Rape'})
            wanted_columns.append('Rape')
        # elif check_name_of_column_in_match(col_name,matches_fraud):
        #     df = df.rename(columns={df.columns[j]: 'Fraud'})
        #     wanted_columns.append('Fraud')
        # elif check_name_of_column_in_match(col_name,matches_rash_driving):
        #     df = df.rename(columns={df.columns[j]: 'Rash driving'})
        #     wanted_columns.append('Rash driving')
        elif check_name_of_column_in_match(col_name,matches_assault_on_women):
            df = df.rename(columns={df.columns[j]: 'Assault on women'})
            wanted_columns.append('Assault on women')
        elif not check_name_of_column_in_match(col_name,matches_not_contain_total_crimes) and check_name_of_column_in_match(col_name,matches_total_crimes):
            df = df.rename(columns={df.columns[j]: 'Total crimes'})
            wanted_columns.append('Total crimes')

# To remove unwanted columns
    df = df.filter(wanted_columns)
    return df
```

```python
for index, df in enumerate(df_list):
    df_list[index]=rename_and_remove_unwanted_column(df)
```

```python
def add_state_column(df):
    unwanted_rows=[]
    for index, data in enumerate(df.iloc[:,0]):
        if str(data).lower().find("state") > -1:
            state= data.replace("State","")
            state= state.replace(":","")
        if str(data).isnumeric():
            df.iloc[index,0]=state.strip()
        else:
            unwanted_rows.append(index)

# To remove unwanted rows
    df.drop(df.index[unwanted_rows], inplace=True)
    
# To rename column
    df.rename(columns={"S. No.": "State"},inplace=True)
    
    return df
```

```python
def add_year_column(df,year):
    df["Year"]=year
    return df
```

```python
for index, df in enumerate(df_list):
    df_list[index]=add_state_column(df)
    df_list[index]=add_year_column(df,year_list[index])
```
